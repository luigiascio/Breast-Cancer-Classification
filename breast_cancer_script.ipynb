# Breast Cancer classification

## 1. Import the dependencies

```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

## 2. Data Collection & Processing

```python
# loading the data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()

#loading in dataframe
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)

data_frame.head()
```

Adding the 'target' column to the data frame, 0 and 1:

```python
data_frame['label'] = breast_cancer_dataset.target
data_frame.tail()
```

Number of rows and columns in the dataset:

```python
# number of rows and columns in the dataset
data_frame.shape
```

Let's get some information about the data:

```python
#getting some information about the data
data_frame.info()
```

Check for missing value:

```python
#checking for missing value
data_frame.isnull().sum()
```

## 3. Statistical measures about the data

The initial statistical analysis reveals a high standard deviation in the 'mean area', indicating increased variability and greater dispersion of data points around the mean. The mean's deviation from the median is notable, likely influenced by the presence of outliers.

```python
#statistical measures about the data
descrizione = data_frame.describe()
mediana = data_frame.median()

# Unisci le statistiche descrittive con la mediana usando pd.concat
data_frame_describe = pd.concat([descrizione, mediana.to_frame().transpose()])
data_frame_describe.index = pd.Index(['count', 'mean', 'std', 'min', '25%', '50%', '75%', 'max', 'median'])
data_frame_describe
```

Checking the distribution of Target Variable:

```python
# checking the distribution of Target Variable
data_frame['label'].value_counts()
# do the percent
```

## 4. ML implementation

Separating the features and target

```python
X = data_frame.drop(columns = 'label', axis = 1)
Y = data_frame['label']
```

Splitting the data into training data & testind data:

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 2)
print(X.shape, X_train.shape, X_test.shape)
```

### 4.1 Model training

It's used "Logistic Regression" 

```python
model = LogisticRegression()
```

```python
#training the logistic regression model using training data

model.fit(X_train, Y_train)
```

### 4.2 Model Evaluation  
Accuracy score:

```python
#accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score (Y_train, X_train_prediction)
print('Accuracy on training data = ', training_data_accuracy)
#very high
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
print('Accuracy on test data = ', test_data_accuracy)
#very high
```
